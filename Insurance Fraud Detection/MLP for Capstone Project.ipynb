# Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np

# Assuming df_new is your DataFrame and 'Classification' is the target column
X = df_new.drop(columns=['Classification'])

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=0.95)  # Retain 95% of variance
X_pca = pca.fit_transform(X_scaled)

# Explained variance ratio
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

# Plotting the explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.show()

# Number of components retained
num_components = X_pca.shape[1]
print(f'Number of components to retain 95% variance: {num_components}')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression  # RFE needs an estimator

# features (X) and target (y)
X = df_new.drop(columns=['Classification']) 
y = df_new['Classification']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Perform Recursive Feature Elimination (RFE) to find the optimum number of features
# Use Logistic Regression as a base estimator for RFE
log_reg = LogisticRegression(max_iter=1000, random_state=0)
rfe = RFE(estimator=log_reg, n_features_to_select=1, step=1)
rfe.fit(X_train_scaled, y_train)

# Rank features based on importance
rfe_ranking = rfe.ranking_
feature_ranking_df = pd.DataFrame({
    'Feature': X.columns,
    'RFE_Rank': rfe_ranking
}).sort_values(by='RFE_Rank')

print("\nFeature ranking based on RFE:\n", feature_ranking_df)

# Retrain MLP Classifier using the top 'n' features (for example, top 10 features)
n_optimal_features = 10  # You can choose based on your preference or the RFE result
X_train_selected = X_train_scaled[:, rfe_ranking <= n_optimal_features]
X_test_selected = X_test_scaled[:, rfe_ranking <= n_optimal_features]

# MLP Classifier model with selected features
mlp_classifier = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=500, random_state=0)
mlp_classifier.fit(X_train_selected, y_train)

# Predict on the testing data
y_test_pred = mlp_classifier.predict(X_test_selected)
y_test_prob = mlp_classifier.predict_proba(X_test_selected)[:, 1]

# Evaluate the model
print("Confusion Matrix (Test Data):")
print(confusion_matrix(y_test, y_test_pred))
print("\nClassification Report (Test Data):")
print(classification_report(y_test, y_test_pred))
print("Testing Accuracy Score: {:.2f}%".format(accuracy_score(y_test, y_test_pred) * 100))
```
