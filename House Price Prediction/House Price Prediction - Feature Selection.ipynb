# Feature Selection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# For feature selection
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
# Displaying all the columns of the dataframe
pd.pandas.set_option('display.max_columns', None)

# Displaying all the rows of the column of the dataframe 
# (As it will show all the rows whether there are thousands or millions)
#pd.pandas.set_option('display.max_rows', None)
```

```python
df = pd.read_csv("C:/Users/Nishant Moona/Downloads/Housing Prices Dataset/X_train.csv")
```

```python
df.head()
```

```python
# Capturing the depenent feature
y_train = df[['SalePrice']]
```

```python
# Dropping dependent feature from the dataset
X_train = df.drop(['Id','SalePrice'], axis = 1)
```

```python
# Apply feature selection
# Specifying the Lasso Regression model and selecting suitable
# alpha (equivanet to penalty).
# Bigger the alpha, less features will be selected

# SelectFromModel object from sklearn will select the features which have coefficients non-zero.
feature_sel_model = SelectFromModel(Lasso(alpha = 0.005,random_state = 0))
feature_sel_model.fit(X_train, y_train)
```

```python
# Values of True and False, count of all independent features.
# True - Feature is important
# False - you can drop that feature
feature_sel_model.get_support()
```

```python
# Printing the number of total and selected features

# We can make a list of selected features
selected_feat = X_train.columns[(feature_sel_model.get_support())]

# Print some stats
print('total features: {}'.format((X_train.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrank to zero: {}'.format(np.sum(feature_sel_model.estimator_.coef_ == 0)))
```
